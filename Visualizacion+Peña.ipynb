```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

# Punto 1 Cargar este archivo en Python. Realizar estadísticas descriptivas básicas

```python
df1 = pd.read_csv('HRDataset_v14.csv') #Cargo el archivo csv como un dataframe con Pandas
```

```python
colNumericas = list(df1.select_dtypes("number").columns)  #Selecciono las columnas que contienen números
df1[colNumericas].describe() #Veo el conjunto y las medidas para el mismo
```

```python
colCategoricas = list(df1.select_dtypes("object").columns) #Selecciono las columnas categoricas
df1[colCategoricas].describe() #Veo el conjunto y las medidas para el mismo
```

# Punto 2 Realizar un histograma con los salarios. Qué rango(s) de salarios son los más populares?

```python
fig, ax = plt.subplots()
ax.hist(df1[colNumericas]['Salary'].values.flatten(), bins=10)
ax.set_title(f"Histograma de los Salarios")
ax.set_ylabel('Frecuencia absoluta')
```

```python
#Acá podemos ver que los salarios mas frecuentes son ente 50000 y 100000
```

# Punto 3 Realizar un gráfico de violín con los salarios, discriminados por género.

```python
ax = sns.catplot(data=df1, kind='violin', x='Sex', y='Salary', split=True)
ax.set(xlabel='Sex',ylabel='Salary',title='Salarios por Género')
```

# Punto 4 Graficar la serie de tiempo correspondiente a la fecha de contratación (DateofHire)


```python
plt.figure()
sns.set(rc={'figure.figsize':(20,20)}) #Cambio el tamaño del gráfico para ver el eje X mejor
ax = sns.lineplot(data=df1, x='DateofHire', y='EmpID', linewidth=1)
ax.set(xlabel='Inicio de contrato', ylabel='Densificación', title='Contratos')
```
